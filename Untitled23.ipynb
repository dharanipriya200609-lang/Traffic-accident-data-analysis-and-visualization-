```python

# ============================================================
# TRAFFIC ACCIDENT DATA ANALYSIS AND VISUALIZATION PROJECT
# ============================================================

# -----------------------------
# 1. IMPORT REQUIRED LIBRARIES
# -----------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib

# -----------------------------
# 2. DATASET CREATION (1000 RECORDS)
# -----------------------------
np.random.seed(42)

records = 1000

data = {
    "Accident_ID": range(1, records + 1),
    "Age": np.random.randint(18, 75, records),
    "Gender": np.random.choice(["Male", "Female"], records),
    "Weather": np.random.choice(["Clear", "Rainy", "Foggy"], records),
    "Road_Type": np.random.choice(["Highway", "City Road", "Village Road"], records),
    "Vehicle_Type": np.random.choice(["Car", "Bike", "Bus", "Truck"], records),
    "Speed_kmph": np.random.randint(20, 120, records),
    "Alcohol_Involved": np.random.choice(["Yes", "No"], records),
    "Seatbelt": np.random.choice(["Yes", "No"], records),
    "Accident_Severity": np.random.choice(
        ["Minor", "Serious", "Fatal"], records, p=[0.5, 0.35, 0.15]
    )
}

df = pd.DataFrame(data)
df.to_csv("traffic_accident_data.csv", index=False)

print("Dataset Created Successfully")
print(df.head())

# -----------------------------
# 3. DATASET DESCRIPTION
# -----------------------------
print("\nDataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nStatistical Summary:")
print(df.describe())

# -----------------------------
# 4. DATA PREPROCESSING
# -----------------------------
label_cols = [
    "Gender", "Weather", "Road_Type",
    "Vehicle_Type", "Alcohol_Involved",
    "Seatbelt", "Accident_Severity"
]

le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])

print("\nLabel Encoding Completed")

# -----------------------------
# 5. EXPLORATORY DATA ANALYSIS (EDA)
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Accident_Severity", data=df)
plt.title("Accident Severity Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df["Speed_kmph"], bins=30, kde=True)
plt.title("Speed Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="Accident_Severity", y="Speed_kmph", data=df)
plt.title("Speed vs Accident Severity")
plt.show()

# -----------------------------
# 6. FEATURE ENGINEERING
# -----------------------------
df["High_Speed"] = df["Speed_kmph"].apply(lambda x: 1 if x > 80 else 0)
df["Risk_Factor"] = df["Alcohol_Involved"] + df["High_Speed"]

print("\nFeature Engineering Completed")

# -----------------------------
# 7. SPLIT DATA
# -----------------------------
X = df.drop(["Accident_ID", "Accident_Severity"], axis=1)
y = df["Accident_Severity"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# -----------------------------
# 8. MODEL BUILDING
# -----------------------------

# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)

# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)

# Random Forest
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# -----------------------------
# 9. MODEL EVALUATION
# -----------------------------
print("\n--- Logistic Regression ---")
print("Accuracy:", accuracy_score(y_test, lr_pred))
print(classification_report(y_test, lr_pred))

print("\n--- Decision Tree ---")
print("Accuracy:", accuracy_score(y_test, dt_pred))
print(classification_report(y_test, dt_pred))

print("\n--- Random Forest ---")
print("Accuracy:", accuracy_score(y_test, rf_pred))
print(classification_report(y_test, rf_pred))

# Confusion Matrix
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, rf_pred), annot=True, fmt="d")
plt.title("Random Forest Confusion Matrix")
plt.show()

# -----------------------------
# 10. SAVE MODEL (DEPLOYMENT)
# -----------------------------
joblib.dump(rf, "traffic_accident_model.pkl")
joblib.dump(scaler, "scaler.pkl")

print("\nModel Saved Successfully")

# -----------------------------
# 11. SAMPLE PREDICTION
# -----------------------------
sample_data = [[
    25,  # Age
    1,   # Gender
    0,   # Weather
    1,   # Road_Type
    0,   # Vehicle_Type
    90,  # Speed
    1,   # Alcohol
    0,   # Seatbelt
    1,   # High Speed
    2    # Risk Factor
]]

sample_scaled = scaler.transform(sample_data)
prediction = rf.predict(sample_scaled)

print("\nPredicted Accident Severity:", prediction)

# ============================================================
# END OF PROJECT
# ============================================================
```
